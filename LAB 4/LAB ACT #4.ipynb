```
import matplotlib.pyplot as plt

class GroceryItem:
    def __init__(self, name, price, quantity, item_type):
        self.name = name
        self.price = price
        self.quantity = quantity
        self.item_type = item_type

    def calculate_sum(self):
        return self.price * self.quantity

    def __str__(self):
        return f"{self.name:8} ({self.item_type:10}) | PHP {self.price:3} | x{self.quantity:2} | Total: PHP {self.calculate_sum():4}"

    def __repr__(self):
        return self.__str__()

class Fruit(GroceryItem):
    def __init__(self, name, price, quantity):
        super().__init__(name, price, quantity, "Fruit")

class Vegetable(GroceryItem):
    def __init__(self, name, price, quantity):
        super().__init__(name, price, quantity, "Vegetable")

def TotalSum(grocery_list):
    return sum(item.calculate_sum() for item in grocery_list)

def plot_grocery_data(grocery_list, title):
    names = [item.name for item in grocery_list]
    totals = [item.calculate_sum() for item in grocery_list]
    colors = ['orange' if isinstance(item, Fruit) else 'green' for item in grocery_list]

    plt.figure(figsize=(10, 6))
    bars = plt.bar(names, totals, color=colors)

    plt.title(title)
    plt.xlabel('Grocery Items')
    plt.ylabel('Total Cost (PHP)')
    plt.grid(axis='y', linestyle='--', alpha=0.7)

    # Add value labels on top of each bar
    for bar in bars:
        height = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2., height,
                 f'PHP {int(height)}',
                 ha='center', va='bottom')

    # Add legend
    fruit_patch = plt.Rectangle((0,0),1,1,fc='orange', edgecolor = 'black')
    veg_patch = plt.Rectangle((0,0),1,1,fc='green', edgecolor = 'black')
    plt.legend([fruit_patch, veg_patch], ['Fruits', 'Vegetables'])

    plt.tight_layout()
    plt.show()

def main():
    # Problem 1 & 2: Create classes and populate grocery list
    GroceryList = [
        Fruit("Apple", 10, 7),
        Fruit("Banana", 10, 8),
        Vegetable("Broccoli", 60, 12),
        Vegetable("Lettuce", 50, 10)
    ]

    # Display all items
    print("="*60)
    print("JENNA'S ORIGINAL GROCERY LIST")
    print("="*60)
    print(f"{'Item':8} {'Type':10} | {'Price':5} | {'Qty':3} | {'Total':6}")
    print("-"*60)
    for item in GroceryList:
        print(item)
    print("-"*60)
    print(f"TOTAL SUM: PHP {TotalSum(GroceryList)}")
    print("="*60)

    # Plot original data
    plot_grocery_data(GroceryList, "Jenna's Original Grocery Expenses")

    # Problem 4: Remove Lettuce
    GroceryList = [item for item in GroceryList if item.name != "Lettuce"]

    # Display updated list
    print("\n" + "="*60)
    print("JENNA'S UPDATED GROCERY LIST (AFTER REMOVING LETTUCE)")
    print("="*60)
    print(f"{'Item':8} {'Type':10} | {'Price':5} | {'Qty':3} | {'Total':6}")
    print("-"*60)
    for item in GroceryList:
        print(item)
    print("-"*60)
    print(f"NEW TOTAL SUM: PHP {TotalSum(GroceryList)}")
    print("="*60)

    # Plot updated data
    plot_grocery_data(GroceryList, "Jenna's Grocery Expenses After Removing Lettuce")

if __name__ == "__main__":
    main()
```
